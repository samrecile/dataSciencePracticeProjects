# Tennis Player Analysis
*In this project, I use a linear regression model to predict tennis player performance based on their playing habits/statistics*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# load and examine the data
data = pd.read_csv('tennis_stats.csv')
print(data.head())
print(len(data))
print(data['Player'].value_counts())
col = data.Wins
print(f'The most wins are {col.max()}')
```

```python
plt.scatter(data.Aces, data.Wins)
plt.title('Aces v Wins')
plt.xlabel('Aces')
plt.ylabel('Wins')
plt.show()
plt.clf()
plt.scatter(data.Ranking, data.Wins)
plt.xlabel('Ranking')
plt.ylabel('Wins')
plt.title('Ranking v Wins')
plt.show()
plt.clf()
plt.scatter(data['FirstServe'], data.Wins)
plt.xlabel('First Serve')
plt.ylabel('Wins')
plt.title('First Serve v Wins')
plt.show()
plt.clf()
plt.scatter(data.ReturnPointsWon, data.Wins)
plt.xlabel('Return Points')
plt.ylabel('Wins')
plt.title('Return Points v Wins')
plt.show()
plt.clf()
```

```python
# scatter first serve return points won to wins
features = data[['FirstServeReturnPointsWon']]
outcome = data[['Wins']]
plt.scatter(features, outcome)
plt.title('Real First Serve Return Points v Wins')
plt.xlabel('First Serve Return Points')
plt.ylabel('Wins')
plt.show()
plt.clf()

# linear model comparing first server points to wins

features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
mlr = LinearRegression()
mlr.fit(features_train, outcome_train)
y_predicted = mlr.predict(features_test)
plt.scatter(y_predicted, outcome_test)
plt.title('Wins v First Serve Return Points')
plt.xlabel('Wins')
plt.ylabel('First Serve Return Points')
plt.show()
print(f'The score for this model is {mlr.score(features_test, outcome_test)}')
plt.clf()
```

```python
# linear model comparing aces to wins
features = data[['Aces']]
outcome = data[['Wins']]
plt.scatter(data['Aces'], data['Wins'])
plt.title('Real Aces v Wins')
plt.xlabel('Aces')
plt.ylabel('Wins')
plt.show()
plt.clf()
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
mlr = LinearRegression()
mlr.fit(features_train, outcome_train)
y_predicted = mlr.predict(features_test)
print(features_test.max())
plt.scatter(y_predicted, outcome_test)
plt.title('Wins v Aces')
plt.xlabel('Aces')
plt.ylabel('Wins')
plt.show()
plt.clf()
print(f'The score for this model is {mlr.score(features_test, outcome_test)}')
```

```python
# scatter first server points to winnings
features = data[['FirstServePointsWon']]
outcome = data[['Winnings']]
plt.scatter(features, outcome)
plt.title('FirstServePointsWon v Winnings')
plt.xlabel('First Serve Points Won')
plt.ylabel('Winnings')
plt.show()
plt.clf()
# linear model comparing first server points to winnings
features = data[['FirstServePointsWon']]
outcome = data[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
mlr = LinearRegression()
mlr.fit(features_train, outcome_train)
print(f'The score for this model is {mlr.score(features_test, outcome_test)}')
y_predicted = mlr.predict(features_test)
plt.scatter(y_predicted, outcome_test)
plt.title('First Serve Points Won v Winnings')
plt.xlabel('First Serve points Won')
plt.ylabel('Winnings')
plt.show()
plt.clf()
```

```python
# multiple linear regression seeing the impact aces & breakPointsConverted have on winnings
features = data[['Aces', 'BreakPointsConverted']]
outcome = data[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
mlr = LinearRegression()
mlr.fit(features_train, outcome_train)
print(f'The score for this model is {mlr.score(features_test, outcome_test)}')
y_predicted = mlr.predict(features_test)
plt.scatter(y_predicted, outcome_test)
plt.title('Aces & BreakPointsConverted v Winnings')
plt.xlabel('Aces & BreakPointsConverted')
plt.ylabel('Winnings')
plt.show()
plt.clf()
```

```python
# multiple linear regression seeing the impact SecondServePointsWon & BreakPointsFaced have on winnings
features = data[['SecondServePointsWon', 'BreakPointsFaced']]
outcome = data[['Winnings']]
features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
mlr = LinearRegression()
mlr.fit(features_train, outcome_train)
print(f'The score for this model is {mlr.score(features_test, outcome_test)}')
y_predicted = mlr.predict(features_test)
plt.scatter(y_predicted, outcome_test)
plt.title('SecondServePointsWon & BreakPointsFaced v Winnings')
plt.xlabel('Winnings')
plt.ylabel('SecondServePointsWon & BreakPointsFaced')
plt.show()
plt.clf()
```

## Conclusion:
1. Categories like aces and break points converted have a significant impact on winning
2. Categories like first serve points won and first serve return points won have a lesser than expected impact on winnings
